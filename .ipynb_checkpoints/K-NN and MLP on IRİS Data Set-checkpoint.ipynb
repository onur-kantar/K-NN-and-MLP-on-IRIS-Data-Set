# KNN

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.metrics import plot_confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

```python
sns.set_style("whitegrid")
```

```python
iris = pd.read_csv("iris.csv")
```

```python
sns.pairplot(iris, hue="Species", palette="husl", markers=["o", "s", "D"], plot_kws={'alpha':0.5});
```

```python
iris.head()
```

```python
x = iris.iloc[:,:4].values #bağımsız değişkenler
y = iris.Species.values #bağımlı değişken
```

```python
print(x[:5])
print(y[:5])
```

```python
x_train, x_test,y_train,y_test = train_test_split(x,y,test_size=0.33, random_state=0)
```

```python
knn = KNeighborsClassifier(n_neighbors=3, metric='minkowski')
knn.fit(x_train,y_train) #eğitim

y_pred = knn.predict(x_test)#test setini dene

cm = plot_confusion_matrix(y_test,y_pred)

f, ax = plt.subplots(figsize =(5,5))
sns.heatmap(cm,annot = True,linewidths=0.5,ax=ax, cbar=False)
plt.xlabel("y_pred")
plt.ylabel("y_true")
plt.show()
```
